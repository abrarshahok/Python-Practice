```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
plt.plot(x, y)
```

```python
plt.scatter(x[::10], y[::10])
```

```python
z = np.cos(x)
plt.scatter(x[::10], y[::10], color='red')
plt.scatter(x[::10], z[::10])
```

```python
plt.scatter(x, y, color='red')
plt.scatter(x, z)
```

```python
plt.plot(x, y, color='red')
plt.plot(x, z)
```
