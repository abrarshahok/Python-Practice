```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
plt.plot(x, y)
```

```python
plt.scatter(x[::10], y[::10])
```

```python
z = np.cos(x)
plt.scatter(x[::10], y[::10], color='red')
plt.scatter(x[::10], z[::10])
```

```python
plt.scatter(x, y, color='red')
plt.scatter(x, z)
```

```python
plt.plot(x, y, color='red')
plt.plot(x, z)
```

# Matplotlib Global API

```python
plt.figure(figsize=(12, 6))
plt.title('My Plot')

plt.subplot(1, 2, 1) # row, column, selected panel
plt.plot(x, x ** 2, label='X^2')
plt.plot([0, 0, 0], [-10, 0, 100], label='Vertical Line') 
plt.legend() # Helps to show the labels
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2), label='-(X^2)')
plt.plot([-10, -0, 10], [-50, -50, -50], label='Horizontal Line')
plt.legend()
plt.xlabel('X')
plt.ylabel('X Squared')
```
