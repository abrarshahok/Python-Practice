```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('covid_19_data.csv')
data
```

```python
data.drop(['SNo','Last Update'],axis = 1, inplace=True)
```

```python
data.rename(columns={'ObservationDate':'Date','Province/State':'Province','Country/Region':'Country'},inplace=True)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.head(10)
```

```python
data.describe()
```

```python
data.info()
```

```python
data = data.fillna('NA')
```

```python
data[data['Province'] == 'NA']
```

```python
imputer = SimpleImputer(strategy='constant')
df = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
df2 = df.groupby('Country')[['Confirmed','Deaths','Recovered']].sum().reindex()
df2.head(10)
```

```python
for country in countries:
    C = df[df['Country'] == country].reset_index()
    plt.plot(np.arange(0,len(C)), C['Confirmed'], color='blue',label='Confirmed')
    plt.plot(np.arange(0,len(C)), C['Recovered'], color='green',label='Recovered')
    plt.plot(np.arange(0,len(C)), C['Deaths'], color='red',label='Deaths')
    plt.title(country)
    plt.xlabel('Days since first suspect')
    plt.xlabel('Number of cases')
    plt.legend()
    plt.show()
```

```python
C = df
plt.plot(np.arange(0,len(C)), C['Confirmed'], color='blue',label='Confirmed')
plt.plot(np.arange(0,len(C)), C['Recovered'], color='green',label='Recovered')
plt.plot(np.arange(0,len(C)), C['Deaths'], color='red',label='Deaths')
plt.title('World')
plt.xlabel('Days since first suspect')
plt.xlabel('Number of cases')
plt.legend()
plt.show()
```
